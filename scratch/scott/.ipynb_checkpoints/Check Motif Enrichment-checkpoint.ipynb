```julia
# http://www.ncbi.nlm.nih.gov/projects/genome/assembly/grc/human/data/
chrLengths = [
    248956422,
    242193529,
    198295559,
    190214555,
    181538259,
    170805979,
    159345973,
    145138636,
    138394717,
    133797422,
    135086622,
    133275309,
    114364328,
    107043718,
    101991189,
    90338345,
    83257441,
    80373285,
    58617616,
    64444167,
    46709983,
    50818468,
    156040895,
    57227415
]

chrOffsets = zeros(Int64, length(chrLengths))
for i in 1:length(chrLengths)
    chrOffsets[i] = sum(chrLengths[1:i-1])
end

function chr_offset(chr)
    if chr == "chr1"
        return chrOffsets[1]
    elseif chr == "chr2"
        return chrOffsets[2]
    elseif chr == "chr3"
        return chrOffsets[3]
    elseif chr == "chr4"
        return chrOffsets[4]
    elseif chr == "chr5"
        return chrOffsets[5]
    elseif chr == "chr6"
        return chrOffsets[6]
    elseif chr == "chr7"
        return chrOffsets[7]
    elseif chr == "chr8"
        return chrOffsets[8]
    elseif chr == "chr9"
        return chrOffsets[9]
    elseif chr == "chr10"
        return chrOffsets[10]
    elseif chr == "chr11"
        return chrOffsets[11]
    elseif chr == "chr12"
        return chrOffsets[12]
    elseif chr == "chr13"
        return chrOffsets[13]
    elseif chr == "chr14"
        return chrOffsets[14]
    elseif chr == "chr15"
        return chrOffsets[15]
    elseif chr == "chr16"
        return chrOffsets[16]
    elseif chr == "chr17"
        return chrOffsets[17]
    elseif chr == "chr18"
        return chrOffsets[18]
    elseif chr == "chr19"
        return chrOffsets[19]
    elseif chr == "chr20"
        return chrOffsets[20]
    elseif chr == "chr21"
        return chrOffsets[21]
    elseif chr == "chr22"
        return chrOffsets[22]
    elseif chr == "chrX"
        return chrOffsets[23]
    elseif chr == "chrY"
        return chrOffsets[24]
    end
end
```

```julia
function parse_chr_pos(line)
    parts = split(line)
    chr_offset(parts[1])+int(parts[2])
end

type PositionStream
    lineItr
    pos
    value
    nextPos
end
function eof(s::PositionStream)
    Base.done(s.lineItr, nothing)
end
function next(s::PositionStream)
    s.pos = s.nextPos
    s.value = 1
    s.nextPos = parse_chr_pos(Base.next(s.lineItr,nothing)[1])
    while s.nextPos == s.pos && !done(s.lineItr, nothing)
        s.value += 1
        s.nextPos = parse_chr_pos(Base.next(s.lineItr,nothing)[1])
    end
end

function open_pstream(cmd)
    lineItr = eachline(cmd)
    pos = parse_chr_pos(Base.next(lineItr,nothing)[1])
    value = 1
    nextPos = parse_chr_pos(Base.next(lineItr,nothing)[1])
    while nextPos == pos && !Base.done(lineItr,nothing)
        value += 1
        nextPos = parse_chr_pos(Base.next(lineItr,nothing)[1])
    end
    PositionStream(lineItr, pos, value, nextPos)
end
```

```julia
function fill_chunk!(dataMatrix, startPos, positionStreams)
    chunkSize = size(dataMatrix)[1]
    for i in 1:length(positionStreams)
        s = positionStreams[i]
        while s.pos-startPos+1 <= chunkSize && !eof(s)
            dataMatrix[s.pos-startPos+1,i] = s.value
            next(s)
        end
    end
end
```

```julia
function smooth_track!(prevBlock, currBlock, nextBlock, srcCol, destCol, windowSize)
    windowSum = 0
    blockSize = size(currBlock)[1]
    for i in -2*windowSize:blockSize
        
        # add new positions to the window
        if i+windowSize <= 0
            windowSum += prevBlock[i+windowSize+end,srcCol]
        elseif i+windowSize <= blockSize
            windowSum += currBlock[i+windowSize,srcCol]
        else
            windowSum += nextBlock[i+windowSize-blockSize,srcCol]
        end
        
        # remove old positions
        if i > -windowSize # make sure we only remove what has been added
            if i-windowSize <= 0
                windowSum -= prevBlock[i-windowSize+end,srcCol]
            elseif i-windowSize <= blockSize
                windowSum -= currBlock[i-windowSize,srcCol]
            else
                windowSum -= nextBlock[i-windowSize-blockSize,srcCol]
            end
        end
        
        # store the current window average
        if i > 0
            currBlock[i,destCol] = windowSum/(2*windowSize)
        end
    end
end
```

```julia
# processorFunc will be passed a block of data at a time
function block_loop(fileIds, processorFunc; chunkSize=10000, chunks=10)
    
    # open process streams from the bam files
    positionStreams = PositionStream[]
    for id in fileIds
        push!(positionStreams, open_pstream(`samtools view -F 0x10 $dataRoot/$id.bam` |> `cut -f 3,4`))
    end

    # read the data block by block
    nIndVars = length(fileIds)-1 
    P = nIndVars*4 + 2
    dataBlocks = Any[]
    push!(dataBlocks, zeros(chunkSize, P))
    push!(dataBlocks, zeros(chunkSize, P))
    push!(dataBlocks, zeros(chunkSize, P))
    dataBlocks[1][:,end] = 1
    dataBlocks[2][:,end] = 1
    dataBlocks[3][:,end] = 1
    pos = 1
    fill_chunk!(dataBlocks[2], pos, positionStreams) # fill curr index
    pos += chunkSize
    fill_chunk!(dataBlocks[3], pos, positionStreams) # fill next index
    pos += chunkSize
    N = 0
    for i in 1:chunks

        # compute what our prev, current, and next indexes are
        prevInd = (i-1)%3+1
        currInd = (i)%3+1
        nextInd = (i+1)%3+1

        # here we add the smoothed track copies
        # note that we need a previous and next block to eliminate the boundary effects of chunking
        for j in 2:length(fileIds) # the first file is the target so don't smooth that
            smooth_track!(dataBlocks[prevInd], dataBlocks[currInd], dataBlocks[nextInd], j, j+nIndVars, 10)
            smooth_track!(dataBlocks[prevInd], dataBlocks[currInd], dataBlocks[nextInd], j, j+2*nIndVars, 100)
            smooth_track!(dataBlocks[prevInd], dataBlocks[currInd], dataBlocks[nextInd], j, j+3*nIndVars, 1000)
        end

        # compute the running C^T*C terms
        processorFunc(dataBlocks[currInd], N+1)
        N += chunkSize

        println("processed $N positions...")

        if i == 1000000# || i == 101
            println("$prevInd $currInd $nextInd")
            println(dataBlocks[currInd])
        end

        # this loads the next chunk of data
        dataBlocks[prevInd][:,1:end-1] = 0 # don't reset the constant column
        fill_chunk!(dataBlocks[prevInd], pos, positionStreams)
        pos += chunkSize

    end

    # clean up the process streams
    for s in positionStreams
        close(s.lineItr.stream)
    end
end
```

<h3>Fit a linear model to the data matrix</h3>

```julia
P = (length(fileIds)-1)*4 + 2
C = zeros(P,P)
function buildC(block, startPos)
    global C
    C .+= block'block
    if startPos%10000000 == 1
        open(f->write(f, C), "/tmp/C_$startPos.dat", "w")
    end
end
block_loop(fileIds, buildC, chunkSize=1000000, chunks=3000)
beta = inv(C[2:end,2:end])*C[2:end,1]
```

```julia
chunkSize = 1000000
chunks = 1
pred = zeros(chunkSize*chunks)
target = zeros(chunkSize*chunks)
function build_pred(block, startPos)
    global pred
    global target
    for i in 1:size(block)[1]
        pred[startPos+i-1] = dot(vec(block[i,2:end]),beta)
        target[startPos+i-1] = block[i,1]
    end
end
block_loop(fileIds, build_pred, chunkSize=chunkSize, chunks=chunks)
```

```julia
function create_forward_track(bamFile)
    N = 100000000
    data = zeros(Float32,N)
    count = 0
    for line in eachline(`samtools view -F 0x10 $bamFile` |> `cut -f 3,4`)
        if line[1] == '0'
            parts = split(line)
            pos = int(parts[3])
            if parts[2] == "chr1" && pos < N
                data[pos] = 1
            else
                break
            end
        end
        
        count += 1
        #if count > 100 break end
        #break
    end
    data
end
function smooth_track!(outData, outDataCol, inData, windowSize)
    windowSum = 0
    for i in 1-windowSize:length(inData)
        if i+windowSize <= length(inData)
            windowSum += inData[i+windowSize]
        end
        if i > windowSize
            windowSum -= inData[i-windowSize]
        end
        if i > 0
            outData[i,outDataCol] = windowSum#sum(data[max(1,i-windowSize):min(length(data),i+windowSize)])
        end
    end
end
targetData = create_forward_track("$dataRoot/ENCFF000NYK.bam")
controlData1 = create_forward_track("$dataRoot/ENCFF000ODO.bam")
controlData2 = create_forward_track("$dataRoot/ENCFF000ODV.bam")
controlData3 = create_forward_track("$dataRoot/ENCFF000ODZ.bam")
controlData4 = create_forward_track("$dataRoot/ENCFF000OEB.bam")
function create_control_data(tracks...)
    numTracks = length(tracks)
    controlData = zeros(Float32, length(tracks[1]), numTracks*4)
    for i in 1:length(tracks)
        ind = (i-1)*4 + 1
        controlData[:,ind] = tracks[i]
        smooth_track!(controlData, ind+1, tracks[i], 10)
        smooth_track!(controlData, ind+2, tracks[i], 100)
        smooth_track!(controlData, ind+3, tracks[i], 1000)
    end
    controlData
end
controlData = create_control_data(controlData1, controlData2, controlData3, controlData4)
size(controlData)
```

```julia
function logistic_loss(target, pred)
    totalLoss = 0.0
    for i in 1:length(target)
        if target[i] == 1
            totalLoss += -log(pred[i])
        else
            totalLoss += -log(1-pred[i])
        end
    end
    totalLoss
end
```

```julia
type ChromosomePosition
    chr::ASCIIString
    pos::Int64
    value::Float64
end

function read_peaks(f, format)
    positions = ChromosomePosition[]
    for line in eachline(f)
        parts = split(line)
        if parts[1][1] == '#' continue end
        
        if format == "narrowPeak"
            push!(positions, ChromosomePosition(parts[1], int((int(parts[2])+int(parts[2]))/2), float(parts[8])))
        elseif format == "FIMO"
            push!(positions, ChromosomePosition(parts[2], int((int(parts[3])+int(parts[4]))/2), float(parts[6])))
        elseif format == "bed2"
            push!(positions, ChromosomePosition(parts[1], int((int(parts[2])+int(parts[3]))/2), float(parts[4])))
        elseif format == "bed"
            push!(positions, ChromosomePosition(parts[1], int((int(parts[2])+int(parts[3]))/2), float(parts[5])))
        end
    end
    sort!(positions, by=x->-x.value)
end
dataDir = "/projects/leelab/data/pure-seq"
peaksCommon = open(f->read_peaks(f, "bed2"), "$dataDir/ENCFF002CHH_hg38.bed")
#peaksMACSB = GZip.open(f->read_peaks(f, "narrowPeak"), "$dataDir/ENCSR000BQS.peaks.narrowPeak.gz")
#peaksMACS = open(f->read_peaks(f, "narrowPeak"), "$dataDir/ENCSR000BQS_peak.narrowPeak")
#peaksMACSA = open(f->read_peaks(f, "bed2"), "$dataDir/test_peaks_hg38.bed")
peaksMACSA = open(f->read_peaks(f, "bed"), "/scratch2/slund1/ENCSR000BQS_local/local_remap2_pooled_q01_summits.bed")
#peaksMACSA = open(f->read_peaks(f, "bed"), "/scratch2/slund1/ENCSR000BQS_local/local_pooled_q001_summits.bed")
peaksMACSB = open(f->read_peaks(f, "bed"), "/scratch2/slund1/ENCSR000BQS_local/local_pooled_q01_summits.bed")
#peaksMACS = open(f->read_peaks(f, "narrowPeak"), "$dataDir/ENCSR000BQS_hg38peaks.narrowPeak")
motifs = open(f->read_peaks(f, "FIMO"), "$dataDir/REST_MA0138.2_FIMO_p6.txt");
```

```julia
function load_blackList_intervals(f)
    intervals = ChromosomeInterval[]
    for line in eachline(f)
        parts = split(line)
        if parts[1][1] == '#' continue end
        
        push!(intervals, ChromosomeInterval (parts[1], int(parts[2]), int(parts[3])))
    end
    intervals
end
blacklist = open(load_blackList_intervals, "$dataDir/wgEncodeDacMapabilityConsensusExcludable_hg38.bed");
```

```julia
function testoverlap(position::ChromosomePosition, intervals::Array{ChromosomeInterval,1})
    for intr in intervals
        if intr.chromosome == position.chr && position.pos <= intr.last && position.pos >= intr.first
            return true
        end
    end
    
    false
end
```

```julia
peaksMACSClean = ChromosomePosition[]
for i in 1:length(peaksMACS)
    if !testoverlap(peaksMACS[i], blacklist)
        push!(peaksMACSClean, peaksMACS[i])
    end
end
length(peaksMACSClean),length(peaksMACS)
```

```julia
function find_distance(position, positions)
    minDist = 100000000
    for i in 1:length(positions)
        if position.chr == positions[i].chr && abs(position.pos - positions[i].pos) < minDist
            minDist = abs(position.pos - positions[i].pos)
        end
    end
    minDist
end
find_distance(peaks[1], motifs)
```
